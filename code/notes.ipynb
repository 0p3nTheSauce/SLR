```python
q = que()
```

```python
ex = q._peak_run('old_runs', 0)
# ex['results']['best_val_loss']
keys1 = ['admin', 'model']
keys2 = ['results', 'best_val_loss']
crits = [
    lambda x: x == 'S3D',
    lambda x: x < 1.6
]

idxs, runs = q.find_runs('old_runs', [keys1, keys2], crits)
print(idxs)
```

```python
head_sum = {
			"model" : "Model",
			"split" : "Split", 
			"exp_no": "Exp",
			"run_id" : "Run ID",
			"config_path": "Config"
		}
runs_info, stats = q.get_runs_info(runs, head_sum)
print(f"  [{'Idx':>3}] {q.run_str(runs_info[0], stats)}")
for i, info in enumerate(runs_info[1:]):
    # Format with padding for alignment
    print(f"  [{i:3d}] {q.run_str(info, stats)}")
```
